## This code is a "Data Generator" designed to simulate a month's worth of vegetable market inspections across Punjab.

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

def generate_telemetry():
    print(">>> INITIALIZING MAGISTRATE TELEMETRY GENERATION...")
    
    cities = {
        'LAHORE': {'lat': 31.5204, 'lon': 74.3587, 'magistrates': range(100, 150)},
        'RAWALPINDI': {'lat': 33.5909, 'lon': 73.0535, 'magistrates': range(200, 230)},
        'FAISALABAD': {'lat': 31.4187, 'lon': 73.0791, 'magistrates': range(300, 340)},
        'MULTAN': {'lat': 30.1978, 'lon': 71.4697, 'magistrates': range(400, 425)}
    }

    start_date = datetime(2026, 1, 20)
    dates = [start_date + timedelta(days=i) for i in range(31)]
    
    commodities = ['Potato', 'Tomato', 'Onion', 'Banana']
    
    records = []
    inspection_id = 1000000

    print(">>> Simulating 30 days of province-wide inspections...")

    for current_date in dates:
        for city, data in cities.items():
            for mag_id in data['magistrates']:
                
                # Default daily profile
                num_inspections = random.randint(5, 15)
                start_hour = random.randint(8, 10)
                current_time = current_date.replace(hour=start_hour, minute=random.randint(0, 59))
                
                # --- ANOMALY INJECTION LOGIC ---
                
                # 1. The Night Owl (Magistrates 105, 205)
                if mag_id in [105, 205]:
                    current_time = current_date.replace(hour=23, minute=random.randint(0, 30))
                
                # 2. The Speedrunner (Magistrates 310, 410)
                is_speedrunner = (mag_id in [310, 410])
                if is_speedrunner: num_inspections = 35
                
                # 3. The Couch Potato (Magistrates 115, 315)
                is_couch_potato = (mag_id in [115, 315])
                fixed_lat = data['lat'] + random.uniform(-0.05, 0.05)
                fixed_lon = data['lon'] + random.uniform(-0.05, 0.05)

                for _ in range(num_inspections):
                    inspection_id += 1
                    
                    # Time increment
                    if is_speedrunner:
                        current_time += timedelta(minutes=random.randint(1, 3)) # ILLEGAL: < 8 mins
                    else:
                        current_time += timedelta(minutes=random.randint(15, 45)) # LEGAL: > 8 mins
                        
                    # Location increment
                    if is_couch_potato:
                        lat, lon = fixed_lat, fixed_lon # ILLEGAL: Not moving
                    else:
                        lat = data['lat'] + random.uniform(-0.05, 0.05)
                        lon = data['lon'] + random.uniform(-0.05, 0.05)
                        
                    # Inspection details
                    commodity = random.choice(commodities)
                    violation_found = random.choices(['Yes', 'No'], weights=[0.15, 0.85])[0]
                    
                    fine_amount = 0
                    fine_status = 'N/A'
                    
                    if violation_found == 'Yes':
                        fine_amount = random.choice([2000, 5000, 10000, 20000])
                        # 4. Revenue Leaker (Magistrates 120, 220)
                        if mag_id in [120, 220]:
                            fine_status = 'Unpaid'
                        else:
                            fine_status = random.choices(['Paid', 'Unpaid'], weights=[0.8, 0.2])[0]
                    
                    records.append({
                        'Inspection_ID': f"INS-{inspection_id}",
                        'Magistrate_ID': f"MAG-{mag_id}",
                        'Date': current_time.strftime('%Y-%m-%d'),
                        'Time': current_time.strftime('%H:%M:%S'),
                        'City': city,
                        'Latitude': round(lat, 6),
                        'Longitude': round(lon, 6),
                        'Commodity_Checked': commodity,
                        'Violation_Found': violation_found,
                        'Fine_Amount': fine_amount,
                        'Fine_Status': fine_status
                    })

    df = pd.DataFrame(records)
    
    # Save the asset
    output_filename = "Magistrate_Telemetry_Log_New.csv"
    df.to_csv(output_filename, index=False)
    
    print("\n" + "="*50)
    print(f"SUCCESS: {len(df)} Inspection Records Synthesized.")
    print("ANOMALIES INJECTED:")
    print("  - Night Owls (105, 205) logging post 10 PM")
    print("  - Speedrunners (310, 410) violating 8-min cooldown")
    print("  - Couch Potatoes (115, 315) spoofing GPS coordinates")
    print("  - Revenue Leakers (120, 220) generating massive unpaid fines")
    print(f"Output File: {output_filename}")
    print("="*50)

if __name__ == "__main__":
    generate_telemetry()
```
